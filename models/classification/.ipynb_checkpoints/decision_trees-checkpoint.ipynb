<h1>Decision Trees</h1>

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
phones_df = pd.read_csv('../../preprocessing/smartphones_preprocessed.csv')
```

```python
phones_df.head()
```

```python
phones_df.isna().any().any()
```

```python
# Define the desired ratio of phones in "low" and "high" bins
desired_ratio = 0.7
min_price = phones_df['price'].min()
max_price = phones_df['price'].max()

# Calculate the threshold price based on the desired ratio
threshold_price = phones_df['price'].quantile(desired_ratio)

# Define the bin edges for low and high based on the threshold
bin_edges = [min_price, threshold_price, max_price]

# Define labels for the bins
labels = ['low', 'high']

# Apply the calculated bins to the "price" column
phones_df['price_category'] = pd.cut(phones_df['price'], bins=bin_edges, labels=labels)
phones_df['price_category'].fillna('high', inplace=True)
phones_df.head()
```

```python
phones_df_encoded = pd.get_dummies(phones_df, columns=['os'], prefix='os')
```

```python
phones_df_encoded.head()
```

```python
# Define X (input features) and Y (target)
X = phones_df_encoded.drop(columns=['price', 'price_category', 'brand_name', 'model', 'processor_brand'])
Y = phones_df_encoded['price_category']
feature_names = X.columns
```

```python
X.describe()
```

```python
Y.unique()
```

```python
# Define custom colors for the bars
colors = ['purple', 'orange']

class_counts = phones_df_encoded["price_category"].value_counts()

# Create a bar chart of the class counts
plt.bar(class_counts.index, class_counts.values, color=colors)
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Instances")


for i, count in enumerate(class_counts.values):
    plt.text(i, count + 5, str(count), ha='center', color='black', fontsize=12)

plt.xticks(rotation=0)  
plt.tight_layout()  
plt.show()
```

```python
# Report function - copied from 
# https://github.com/MATF-istrazivanje-podataka-1/materijali_2022-2023/blob/main/05/Najblizi_susedi.ipynb

from termcolor import colored
from sklearn.metrics import classification_report

def report(model, x, y, text = "training"):
    y_pred = model.predict(x)
    
    print(colored("Classification report for model {} on {} data".format(type(model).__name__, text), "green"))
    print("---------------------------------------------------------------------------------")
    print(classification_report(y, y_pred))
    print("---------------------------------------------------------------------------------")
    
    print(colored("Confusion matrix for model {} on {} data ".format(type(model).__name__, text), "green"))
    print("---------------------------------------------------------------------------------")
    print(pd.DataFrame(confusion_matrix(y, y_pred), columns=['high', 'low'], index=['high', 'low']))
    print("---------------------------------------------------------------------------------")
    
```

```python
# Importing the training and testing sets from the KNN file - fpr comparison later.
import pickle

# Loading the data from the file
with open('data.pkl', 'rb') as f:
    X_train, X_test, Y_train, Y_test = pickle.load(f)
```

```python
# Initializing a decision tree classifier
dtc = DecisionTreeClassifier(random_state=41)

# Training the classifier on the training set
dtc.fit(X_train, Y_train)
```

```python
# Generating predictions on the test set
Y_pred = dtc.predict(X_test)

# Visualizing the decision tree
plt.figure()
plot_tree(dtc, feature_names=feature_names, class_names=labels, filled=True)
plt.show()
```

```python
pd.Series(dtc.feature_importances_, index=X.columns).plot.barh()
plt.title("Feature importance")
plt.show()
```

```python
report(dtc, X_train, Y_train)
```

```python
#DOSLO DO PREPRILAGODJAVANJA
```

```python
report(dtc, X_test, Y_test, "test")
```

```python
import seaborn as sns
# Creating a confusion matrix
cm = confusion_matrix(Y_test, Y_pred, labels=labels)

# Creating a heatmap of the confusion matrix
sns.heatmap(cm, annot=True, cmap='Greens', fmt='d', xticklabels=labels, yticklabels=labels)

plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix - Decision Tree Classifier')

plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

params = {'criterion': ['gini', 'entropy'],
    'max_depth': [2, 4, 6, 8],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_leaf_nodes': [None, 5, 10, 20],
    'min_impurity_decrease': [0.0, 0.1, 0.2],
    'ccp_alpha': [0.0, 0.01, 0.1]
         }
```

```python
estimator = GridSearchCV(DecisionTreeClassifier(), param_grid=params, cv=3) # Po defaultu je cv=5
```

```python
estimator.fit(X_train, Y_train)
```

```python
estimator.best_params_
```

```python
estimator.best_score_
```

```python
Y_pred = estimator.best_estimator_.predict(X_test)
```

```python
report(estimator.best_estimator_, X_train, Y_train)
```

```python
report(estimator.best_estimator_, X_test, Y_test, "test")
```

```python
estimator.cv_results_
```

```python
plot_tree(estimator.best_estimator_, feature_names=feature_names, class_names=labels, filled=True)
```

```python
# Creating a confusion matrix

cm = confusion_matrix(Y_test, Y_pred, labels=labels)

# Creating a heatmap of the confusion matrix
sns.heatmap(cm, annot=True, cmap='Greens', fmt='d', xticklabels=labels, yticklabels=labels)


plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix - GridSearchCV')

plt.show()
```

Slucajne sume

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(n_estimators=2)
random_forest.fit(X_train, Y_train)

report(random_forest, X_test, Y_test, "test")

for i in range(1, len(random_forest.estimators_) + 1):
    plt.subplot(1, 2, i)
    plot_tree(random_forest.estimators_[i-1], filled=True)
```

```python
# Kreiranje RandomForestClassifier sa dodatnim parametrom max_depth
random_forest = RandomForestClassifier(n_estimators=2, max_depth=3, random_state=42)
random_forest.fit(X_train, Y_train)

# Prikaz izveštaja o klasifikaciji
y_pred = random_forest.predict(X_test)
print(classification_report(Y_test, y_pred))

# Prikaz stabala odlučivanja
plt.figure(figsize=(12, 6))
for i in range(1, len(random_forest.estimators_) + 1):
    plt.subplot(1, 2, i)
    plot_tree(random_forest.estimators_[i-1], filled=True, feature_names=feature_names, class_names=random_forest.classes_)
    plt.title(f"Tree {i}")
    
plt.tight_layout()
plt.show()
```

```python
models = [dtc, estimator, random_forest]
model_names = ["DecisionTree", "GridSearchCV", "RandomForest"]
colors = ["blue", "red", "green"]
```

```python
from sklearn.metrics import roc_curve, auc
# Choose Low-rated as the positive class
pos_class = "high" 

# Combine the other two classes into a single negative class
neg_class = "low"

# Convert labels to binary format
y_train_bin = (Y_train == pos_class).astype(int)
y_test_bin = (Y_test == pos_class).astype(int)

plt.figure()
for model, model_name, color in zip(models, model_names, colors):
    # Train a classifier on the training data
    model.fit(X_train, y_train_bin)
    
    # Make predictions on the test data
    y_pred = model.predict_proba(X_test)[:, 1]
    
    # Compute ROC curve and area under the curve (AUC)
    fpr, tpr, _ = roc_curve(y_test_bin, y_pred)
    roc_auc = auc(fpr, tpr)
    
    # Plot the ROC curve
    plt.plot(fpr, tpr, color=color, lw=2, label= model_name +'(AUC = %0.2f)' % roc_auc)
    

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - High price')
plt.legend(loc="lower right")
plt.show()
```
