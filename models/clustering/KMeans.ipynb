<h1>K means</h1>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
phones_df = pd.read_csv('../../preprocessing/smartphones_preprocessed.csv')
```

```python
phones_df
```

```python
# Creating a value mapping dictionary
os_mapping = {'ios': 1, 'android': 2, 'other': 3}

# Using the mapping on the 'os' column
phones_df['os'] = phones_df['os'].map(os_mapping)
```

```python
# Define the desired ratio of phones in "low" and "high" bins
desired_ratio = 0.5
min_price = phones_df['price'].min()
max_price = phones_df['price'].max()

# Calculate the threshold price based on the desired ratio
threshold_price = phones_df['price'].quantile(desired_ratio)

# Define the bin edges for low and high based on the threshold
bin_edges = [min_price, threshold_price, max_price]

# Define labels for the bins
labels = ['low', 'high']

# Apply the calculated bins to the "price" column
phones_df['price_category'] = pd.cut(phones_df['price'], bins=bin_edges, labels=labels)
phones_df['price_category'].fillna('high', inplace=True)
phones_df.head()
```

```python
# Getting a list of the first 15 brands
top_15_brands = phones_df['brand_name'].value_counts().index[:15].tolist()

# Mapping brands according to requirements
phones_df['brand_name'] = phones_df['brand_name'].apply(lambda x: x if x in top_15_brands else 'other')
```

```python
# List of brand names
brand_names = ['apple', 'google', 'honor', 'huawei', 'infinix', 'iqoo', 'motorola', 'oneplus',
               'oppo', 'other', 'poco', 'realme', 'samsung', 'tecno', 'vivo', 'xiaomi']

# Creating a map to map brands to numbers
brand_mapping = {brand: index + 1 for index, brand in enumerate(brand_names)}

# Mapping brands to numerical values
phones_df['brand_name'] = phones_df['brand_name'].map(brand_mapping)
```

```python
from sklearn.cluster import KMeans
```

```python
features = [
    'brand_name', 'price', 'avg_rating', '5G_or_not', 'num_cores', 'processor_speed',
    'battery_capacity', 'fast_charging_available', 'fast_charging', 'ram_capacity',
    'internal_memory', 'screen_size', 'refresh_rate', 'num_rear_cameras',
    'os', 'primary_camera_rear', 'primary_camera_front',
    'extended_memory_available', 'resolution_height', 'resolution_width'
]
```

```python
X = phones_df[features]
X
```

```python
# Calculate the Z-score for each numerical column
z_scores = np.abs((X - X.mean()) / X.std())


z_score_threshold = 2.5


outliers = z_scores > z_score_threshold

for column in X.columns:
    column_outliers = X[outliers[column]]  
    print(f"Outliers for column '{column}':")
    print(column_outliers)
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize the features
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=features)
```

```python
num_clusters = 3  # Defining the number of clusters
kmeans = KMeans(n_clusters=num_clusters, n_init=10, random_state=42)
kmeans.fit(X)
```

```python
# Get the cluster labels for each title
cluster_labels = kmeans.labels_

# Add cluster labels to the titles dataset
phones_df['cluster_label'] = cluster_labels
```

```python
print(phones_df[kmeans.labels_== 0].shape)
print(phones_df[kmeans.labels_== 1].shape)
print(phones_df[kmeans.labels_== 2].shape)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(16, 4))

# Extract instances for each cluster
cluster0 = phones_df[kmeans.labels_ == 0]
cluster1 = phones_df[kmeans.labels_ == 1]
cluster2 = phones_df[kmeans.labels_ == 2]

# Count occurrences of 'low' and 'high' in the 'price_category' attribute for each cluster
cluster0_price_counts = cluster0['price_category'].value_counts()
cluster1_price_counts = cluster1['price_category'].value_counts()
cluster2_price_counts = cluster2['price_category'].value_counts()

# Extract labels and frequencies for 'low' and 'high'
labels = ['low', 'high']
colors = sns.color_palette('pastel')
explode = [0.1, 0]

# Plot pie charts for each cluster
axs[0].pie(cluster0_price_counts, labels=labels, autopct='%1.1f%%', explode=explode, colors=colors)
axs[0].set_title("Cluster 1")

axs[1].pie(cluster1_price_counts, labels=labels, autopct='%1.1f%%', explode=explode, colors=colors)
axs[1].set_title("Cluster 2")

axs[2].pie(cluster2_price_counts, labels=labels, autopct='%1.1f%%', explode=explode, colors=colors)
axs[2].set_title("Cluster 3")

plt.savefig('k_means_price_category.png')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# List of cluster numbers for which you want to calculate Silhouette scores
num_clusters = [2, 3, 4, 5, 6, 7, 8, 9, 10]

# List to store Silhouette scores for each cluster number
silhouette_scores = []

# Loop for calculating Silhouette scores for each cluster number
for num in num_clusters:
    model = KMeans(n_clusters=num, n_init=10, random_state=42)
    model.fit(X)
    silhouette_avg = silhouette_score(X, model.labels_)
    silhouette_scores.append(silhouette_avg)


plt.figure(figsize=(10, 5))
plt.plot(num_clusters, silhouette_scores, marker='o', linestyle='-')
plt.xlabel('Broj klastera')
plt.ylabel('Silhouette Skor')
plt.title('Silhouette Skor za Različite Vrednosti Broja Klastera')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = X.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Korelaciona Matrica")
plt.show()
```
